```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import mplleaflet
import seaborn as sns

import prettyplotlib as ppl

import imp
pd.set_option("max_columns", None)
sns.set_style('whitegrid')

%load_ext watermark
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%watermark -v -d -a 'Mario Javier Carrillo' -p matplotlib,numpy,pandas
```

```python
sfcrime = pd.read_pickle('../data/sfcrime.pkl')
sfcrime['Category'].replace("LARCENY/THEFT","LARCENY_THEFT", inplace=True )
sfcrime = sfcrime.rename(columns={'Y': 'latitude', 'X': 'longitude'})
```

```python
plotter = imp.load_source('plotter', '../utils/plotting/maps_visual.py')
from plotter import sampling_data_points
```

```python
frame_1, frame_2 = sampling_data_points(sfcrime, col_cat="Category", category='DRUNKENNESS', col_time='year', 
              num_1=2016, num_2=2017, num_sample=1000)
fig = plt.figure(figsize=(20,15))
ax = fig.gca()
ax = plt.scatter(frame_1['longitude'], frame_1['latitude'], marker="o", color="r", s=20, linewidths=1)
ax = plt.scatter(frame_2['longitude'], frame_2['latitude'], marker="x", color="black", s=20, linewidths=1)
mplleaflet.display(fig=ax.figure)
```
